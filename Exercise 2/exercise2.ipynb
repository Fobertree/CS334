```python
import seaborn as sns
```

```python
plt = sns.scatterplot(df,x='petal_width',y='petal_length', hue='variety')
plt.set_title('Petal Length vs Petal Width (Virginica vs Setosa)')
plt.set_ylabel("Petal Length")
plt.set_xlabel("Petal Width")
```

```python
plt = sns.boxplot(df, x='variety',y='sepal_length')
plt.set_title('Sepal Length By Variety Box Plot')
plt.set_xlabel("Variety")
plt.set_ylabel("Sepal Length")
```
